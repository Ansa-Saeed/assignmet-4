```python
#  question no 1
# Make a calculator using Python with addition , subtraction ,multiplication ,division and power.
 
def add(num1, num2):
    return num1+num2
def sub(num1,num2):
    return num1-num2
def mul(num1, num2):
    return num1*num2
def div(num1, num2):
    return num1/num2
# while True:
choice=input("Enter a Choice 1 to 4: ")
if choice in ('1','2','3','4'):
        x=eval(input("Enter First Number: "))
        y=eval(input("Enter Second Number: "))
        if choice=='1':
            print(x," + ",y,"= ",add(x,y))
        elif choice=='2':
            print(x," - ",y," = ",sub(x,y))
        elif choice=='3':
            print(x," * ",y,"= ",mul(x,y))
        elif choice=='4':
            print(x," \ ",y,"= ",div(x,y))
        else:
            print("no ")
else:
        print("invalid ")
    
```

```python
#  question no 2
# Write a program to check if there is any numeric value in list using for loop.
list1=["hii",23,"helloo","bye",40,50]
for x in list1:
    if type(x)==int:
        print("Contains numeric value In list",x)
    else:
        print("Not contains numeric value In list",x)

```

```python
#  question no 5
#  Write a program to identify duplicate values from list
from collections import Counter
list2=[2,4,4,3,5,6,7,4,2,7,8,9,9,10,23,45,67,67]
d=Counter(list2)
dup_list=list(num for num in d if d[num]>1)
print("Duplicate Values: ",dup_list)
```
